# Section 10 딕셔너리(dict)

**key-value**쌍으로 이루어진 연관 배열
**key를 통해 값 검색 가능**

**{ }** 중괄호 를 통해 정의

**key-value값은 콜론(:)으로 구분, key값은 고유값으로 중복불가**

.keys(), values(), items()로 목록 확인가능 cf) ()가 있어야 호출이 됨. ()는 호출기. ()가 없으면 method 객체 그자체를 의미하게 됨

> my_dict = {'a': 1, 'b': 2}

> print(my_dict.keys)     # <built-in method keys of dict object at 0x...>

> print(my_dict.keys())   # dict_keys(['a', 'b'])

list()로 dict를 list로 변환할 수 있다.
기본은 keys를 list로 반환하되, value로 하고 싶으면 .values() method사용, 둘다하고싶으면 items() method 사용



```python
#예시
dictbox = {'name':'쿼카', 'level':5}
print(dictbox)

#type
print("type은",type(dictbox))

#key로 value찾기
print("name:",dictbox['name'])
print("level:",dictbox['level'])

#key의 value 변경
dictbox['level']=6
print("변경된 level:",dictbox['level'])

#key만 확인
print("key 목록:",dictbox.keys())

#value만 확인
print("value 목록:", dictbox.values())

#key와 value 확인
print("item 목록:", dictbox.items())
```

# Section 11 인덱싱과 슬라이싱
인덱싱: 리스트에서 특정한 '하나의 값'에 접근

슬라이싱: 리스트의 특정 '구간'을 추출

--> 리스트, 문자열처럼 순서가 있는 자료형에 사용가능

단, dict는 순서가 없으므로 사용 불가,

```python
#인덱싱(0부터시작)
listbox = [2,4,6,8,10]

print(listbox[0]) #인덱스 0, 첫번째 값
print(listbox[4]) #인덱스 3, 네번째 값
print(listbox[-1]) #인덱스 -1, 마지막 값
```

```python
#슬라이싱
print(listbox[1:4]) #인덱스 1~3, 이때는 시작지점부터 마지막 지점 '앞'까지만 슬라이싱 됨
#[x:y]일때 x이상 y미만임
#[x:y:z] x이상 y미만 z간격
#끝 또는 시작지점 생략가능
print(listbox[3:]) # 3부터 끝까지
print(listbox[:3]) # 2번인덱스까지
print(listbox[::2]) # 처음부터 간격 2로 끝까지
print(listbox[:]) # 처음부터 끝까지
```

```python
#슬라이싱으로 값변경
listbox[1:3] = [10,20] #1번 부터 2번인덱스를 10,20으로 변경
listbox
```

```python
# enumarte()

listbox=['나는', '장준영', '빅분기', '공부중']

for index, item in enumerate(listbox):
  print(index, item)
```

```python
# zip()

person_info = {
    'name': '사랑',
    'age':20,
    'city': '부산',
    'hobbies':['연애','수영','코딩']
}

for key, value in zip(person_info.keys(), person_info.values()):
  print(key, value)

names = ['준영', '영서', '창혁']
personalities = ['좋음', '보통','나쁨']
grades = ['A', 'B', 'C']

for name, personality, grade in zip(names, personalities, grades):
  print(name, personality, grade)

for index, (name, personality, grade) in enumerate(zip(names, personalities, grades)):
  print(index+ 1, name, personality, grade)

```

```python
#return

def plus(x,y):
  x+y
  result = x+y
  return(result)

a= plus(2,3)
print(a)
```

```python
#여러개의 return값

def min_max(data):
  mi = min(data)
  ma = max(data)
  return mi, ma

listbox = [12, 26, 90, 13, 5]
a,b = min_max(listbox)
print(a,b)
```

```python
def min_max(data):
  mi = min(data)
  ma = max(data)
  return mi, ma

listbox = [12, 26, 90, 13, 5]
b = min_max(listbox) # 밭는개 하나면 튜플형태로 받는다
print(b)
```

```python
#간단한 경우 return뒤에 바로 수식을 지정하기도 함

listbox = [12, 26, 90, 13, 5]

def mean(data):
  return sum(data)/len(data)

mean(listbox)
```
