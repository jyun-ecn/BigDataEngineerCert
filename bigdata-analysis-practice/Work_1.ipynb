# Section 16 이상치, IQR

1. 'views'컬럼의 1사분위수, 3사분위수 그리고 IQR을 계산하시오.
2. 이상치 조건에 맞는 데이터를 찾으시오. (이상치는 1사분위수 - IQR * 1.5 보다 작은 값과 3사분위수 + IQR * 1.5 보다 큰값)
3. 이상치 데이터의 views 컬럼 합을 정수로 구하시오.                          

```python
#데이터 불러오기
import pandas as pd
df = pd.read_csv('type1_data1.csv')

#1사분위수
q1 = df['views'].quantile(.25)
#3사분위수
q3= df['views'].quantile(.75)
#값확인
print("Q1 :",q1)
print("Q3 :",q3)
#IQR
IQR = q3 - q1
print("IQR:",IQR)

#이상치 조건에 맞는 데이터를 찾기
cond = (df['views'] < q1 - IQR *1.5 ) | (df['views'] > q3 + IQR * 1.5)
df = df[cond] # 조건에 맞는 데이터 저장

#이상치 데이터의 합
print("Answer:",int(df['views'].sum()))
```

2. 시간차가 0일인 사용자를 제외하고, 나머지 사용자들의 평균 시간 간격(일 단위)을 계산하시오.
3. 평균 시간 간격보다 긴 시간 간격을 가진 사용자의 수를 정수로 수하시오.

```python
import pandas as pd
df = pd.read_csv('delivery_time.csv')

#datetime
df['주문시간'] = pd.to_datetime(df['주문시간'])

#첫주문 마지막 주문 추출
First_order = df.groupby('user')['주문시간'].min()
Last_order = df.groupby('user')['주문시간'].max()
#시간 간격 계산
gap =(Last_order - First_order).dt.days

#시간차 0일인 사용자 제외
cond = gap>0
gap =gap[cond]

#평균 시간 간격 계산
gap_mean = gap.mean()

#평균 시간 간격보다 긴 시간 간격을 가진 사용자의 수
len(gap[gap > gap_mean])
```

# Section 27 시간 범위, 속도(km/h)
1. 점심시간(10시부터 13시 전까지)에 주문된 배달 데이터를 찾으시오.
2. 점심시간 주문 건 중 과속(평균 속도가 50km/h 이상)하는 주문 수를 정수로 구하시오.
  - 배달시간 = 실제도착시간 - 주문시간
  - 속도(km/h) = 거리(km)/ 시간(h)

```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv('delivery_time.csv')

# pd.to_datetime
df[['주문시간','실제도착시간']] = df[['주문시간', '실제도착시간']].apply(pd.to_datetime)

#hour로 변환
# df['주문시간'] = df['주문시간'].dt.hour << 새로운 col로 대치해야함. 이런식 나중에 실제도착시간과 연산이 안됨

#hour로 조건 설정 후 점신시간 데이터로 제한
cond = (df['주문시간'].dt.hour >=10) & (df['주문시간'].dt.hour < 13)
df = df[cond]

#배달시간 구하기
df['배달시간'] = df['실제도착시간'] - df['주문시간']

#배달시간을 h단위로 변환
df['배달시간'] = df['배달시간'].dt.total_seconds()/60/60

#속도 계산
df['속도'] = df['거리'] / df['배달시간']

#속도가 50km/h이상인 수를 정수로 구하기
cond2 = df['속도'] > 50
df = df[cond2]

len(df)
```

# Section 29 함수, 월별 집계

1. 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산하시오.
2. 월별로 배달료의 총합을 집계하시오.
3. 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 구하시오.

[배달료 기준표]
- 5km 미만: 2,000원
- 5km 이상 ~ 10km 미만: 4,000원
- 10km 이상 ~ 15km 미만: 6,000원
- 15km 이상 ~ 20km 미만: 8,000원

```python
#데이터 불러오기
import pandas as pd
df = pd.read_csv('delivery_time.csv')

#배달료 계산 함수 지정
def delivery_fee(x):
  if x < 5:
    return 2000
  elif x<10:
    return 4000
  elif x<15:
    return 6000
  else:
    return 8000

#배달료 col 추가
df['배달료'] = df['거리'].apply(delivery_fee)

#datetime 변환
df['주문시간'] = pd.to_datetime(df['주문시간'])

#월별로 주문시간 변환
df['주문시간'] = df['주문시간'].dt.to_period('M')

#월별 배달료 총합 집계
df = df.groupby('주문시간')['배달료'].sum()

#배달료가 가장 많은 달 선택
best_month = df.idxmax()

#총 배달료 정수 출력
int(df[best_month])
```
