# Section 1 이진 분류 평가지표(정확도, 정밀도, 재현율, F1score, ROC-AUC)

```python
#이진 분류 데이터
import pandas as pd
y_true = pd.DataFrame([1,1,1,0,0,1,1,1,1,0]) #실제값
y_pred = pd.DataFrame([1,0,1,1,0,0,0,1,1,0]) #예측 값

y_true_str = pd.DataFrame(['A', 'A', 'A','B','B','A','A','A','A','B']) #실제 값
y_pred_str = pd.DataFrame(['A','B','A','A','B','B','B','A','A','B']) #예측 값
```

# 정확도

전체 데이터 중 올바르게 예측된 데이터의 비율

```python
#정확도
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true, y_pred)
print('Accuracy:',accuracy)

accuracy = accuracy_score(y_true_str,y_pred_str)
print('Accuracy:' , accuracy)
```

# 정밀도

예측 참 중 진짜 참예측

```python
from sklearn.metrics import precision_score
precision = precision_score(y_true,y_pred)
print('정밀도:',precision)

precision = precision_score(y_true_str,y_pred_str, pos_label = 'A') #str은 pos_label 지정필요
print('정밀도;', precision)
```

# 재현율

실제 참 중 진짜 참예측

```python
from sklearn.metrics import recall_score
recall = recall_score(y_true, y_pred)
print('재현율:', recall)

recall = recall_score(y_true_str, y_pred_str, pos_label='A')
print('재현율:', recall)
```

# F1 스코어
정밀도와 재현율의 조화 평균, 불균형 데이터에 유용

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_true, y_pred)
print('F1 스코어:', f1)

f1 = f1_score(y_true_str, y_pred_str, pos_label='A')
print('F1 스코어:', f1)
```

# 회귀 평가지표

오차를 설명하는 것이므로 0에 가까울수록 좋다

```python
#회귀 데이터
import pandas as pd
y_true = pd.DataFrame([1,2,5,2,4,4,7,9]) #실제 값
y_pred = pd.DataFrame([1.14,2.53,4.87,3.08,4.21,5.53,7.51,10.32]) #예측값
```

```python
#MSE 실제 값과 예측값의 차이 제곱한 후 평균
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true,y_pred)
mse
```

```python
#MAE Mean Absoulute Error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true,y_pred)
mae
```

```python
# R-squared 결정계수
from sklearn.metrics import r2_score
r2= r2_score(y_true,y_pred)
r2
```

```python
#RMSE(Root Mean Squared Error)
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_true,y_pred)
rmse
```

```python
# MSLE(Mean Squared Log Error) : 작은오차에 큰 가중치
from sklearn.metrics import mean_squared_log_error
msle = mean_squared_log_error(y_true,y_pred)
msle
```

```python
# RMSLE(Root Mean Squared Log Error) : 작은 오차에 큰 가중치
from sklearn.metrics import root_mean_squared_log_error
rmsle = root_mean_squared_log_error(y_true,y_pred)
rmsle
```

```python
# MAPE(Mean Absolute Percentage Error)
mape = (abs(((y_true-y_pred) /y_true)).mean()) * 100
mape
```
