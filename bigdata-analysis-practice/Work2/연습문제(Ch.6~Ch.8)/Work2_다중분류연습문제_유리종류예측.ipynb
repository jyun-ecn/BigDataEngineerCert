- 유리 식별 데이터에서 유리의 종류를 예측하시오.
 - 제공된 데이터 목록: glass_train.csv, glass_test.csv
 - 예측할 컬럼: Type(1,2,3,5,6,7)
- 학습용 데이터(train.csv)를 이용해 약물 종류를 예측하는 모델을 만든 후 이를 평가용 데이터(test.csv)에 적용해 얻은 예측값을 다음과 같은 형식의 CSV 파일로 생성하시오.

제출 파일은 다음 1개의 컬럼을 포함해야 한다.
- pred: 예측값
- 제출 파일명: 'result.csv'
제출한 모델의 성능은 f1-weighted 평가지표에 따라 채점한다.

```python
# 데이터 불러오기
import pandas as pd
train = pd.read_csv('glass_train.csv')
test = pd.read_csv('glass_test.csv')
```

```python
# EDA
print('===== 데이터 크기 =====')
print(train.shape, test.shape)

print('\n===== 자료형 =====')
print(train.info())

print('\n===== 결측치(train) =====')
print(train.isnull().sum())

print('\n===== 결측치(test) =====')
print(test.isnull().sum())
```

```python
# 전처리 생략

# target 추출
target = train.pop('Type')

# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size = 0.2,
    random_state=0
)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
# 모델 학습
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)

rf.fit(X_train, y_train)
pred = rf.predict(X_val)
pred[:5]
```

```python
# 채점
from sklearn.metrics import f1_score
f1 = f1_score(y_val, pred, average='weighted')
f1
```

```python
# randomforest hyperparameter 조정

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size = 0.2,
    random_state=0
)

depths = [3,5,7,9]
estimators = [200,300,400,500]
best_depth = None
best_estimator= None
best_f1 = 0

for depth in depths:
  for estimator in estimators:
    rf = RandomForestClassifier(random_state=0, max_depth=depth, n_estimators = estimator)
    rf.fit(X_train, y_train)
    pred = rf.predict(X_val)
    f1 = f1_score(y_val,pred, average = 'weighted')
    print('f1_socre of', 'depth:',depth, 'and', 'estimator:',estimator,'is', f1)
    if f1 > best_f1:
      best_f1 = f1
      best_depth = depth
      best_estimator = estimator
print('best combination is', 'depth:',best_depth, 'estimator:', best_estimator,)
```

```python
# 최종모델선정
model = RandomForestClassifier(random_state=0, max_depth = 5, n_estimators = 200)
model.fit(X_train,y_train)
pred = model.predict(test)
pred[:5]
```

```python
# 최종파일생성
submit = pd.DataFrame({'pred':pred})
submit.to_csv('result.csv',index=False)
pd.read_csv('result.csv')
```
