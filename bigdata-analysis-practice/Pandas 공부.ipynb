# Section 1 데이터프레임과 시리즈

판다스의 데이터는 **시리즈**, **DataFrame**으로 구성되어 있음

데이터프레임: 행(0)과 열(1)로 구성되어 있는 2차원(표) 형태

import pandas as pd의 형태로 import한다.

시리즈는 pd.series로 데이터를 만들수 있다.

ex.

menu = pd.Series(['비빔밥', '김치찌개', '된장찌개'])  #dtype = object(문자열)
price = pd.Series([10000,9000,8000]) #dtype  = int64

> 대문자 주의, Series는 인덱스도 같이 출력,

**데이터프레임**
pd.DataFrame({
  "칼럼명":데이터,
  "칼럼명":데이터,
})
의 형태로 만듬

df["컬럼명"]으로 특정 칼럼명만 나타낼 수 있다. 1개의 칼럼명만 선택하면 시리즈 형태로 출력한다. dataframe형태로 하려면 여러칼럼할때처럼 리스트의 형태로 작성하면 된다.

** 만약 여러 칼럼을 하려면 [ ]안의 칼럼명을 리스트의 형태로 작성해야한다. **



> index=["A", "B", "C", "D"]처럼 데이터프레임을 지정할때 {}뒤에 ,로 인덱스명을 지정할 수도 있다.




```python
#import
import pandas as pd
```

```python
#Series
menu = pd.Series(['비빔밥', '김치찌개', '된장찌개'])
print(menu)
price = pd.Series([10000,9000,8000])
print(price)
```

```python
#dataframe
#Series로 만들기
df1 = pd.DataFrame({
    "메뉴":menu,
    "가격":price
})

df2 = pd.DataFrame({
    "메뉴": ['햄버거','라면','과자'],
    "가격": [5000,4000,2000],
    "원산지": ["미국", "한국", "중국"]
})

print(df1)
print(df2)
```

```python
#df의 특정 컬럼명 출력
print(df1["가격"]) #시리즈 형태로 반환
print(df2[["메뉴","원산지"]]) #리스트로 입력 후 df형태로 변환
print(df1[["가격"]]) #단일 칼럼명도 마찬가지로 df형태를 원할때

cols =["메뉴","가격"] #이처럼 리스트로 미리 지정후 출력도 가능
print(df2[cols])
```

```python
# type들 확인
# 문제에 특정 유형만 필요한경우가 있으므로, 잘 확인. 특히 하나의 칼럼을 시리즈가아닌 df가 필요한 경우도 다수

print("df      :", type(df2))
print("df[가격]:", type(df2["가격"]))
print("df[['메뉴','가격']]", type(df2[["메뉴","가격"]]))
```

# Section 3 탐색적 데이터 분석(EDA)

  전처리 하기 이전에 데이터에 대한 이해를 위해 결측치, 이상치, 패턴, 변수 파악 등을 하는 과정

1. 샘플 확인

  - head(n): N개의 상위샘플 확인 (기본값 5개)
  - tail(n): n개의 하위샘플 확인
  - sample(n): 임의의 n개 샘플링 기본값은 1개

2. 전체 데이터 크기 확인

  - df.shape :괄호가 없다. 이는 메소드가아닌 속성이라 그렇다. 행의 개수와 열의 개수를 반환

3. 칼럼별 type()확인

  - df.info()

4. 상관관계 확인

  - df.corr(numeric_only=True) 반드시반드시 numeric_only = True를 작성해야 한다. 이는 데이터프레임에 숫자형과 문자형 컬럼이 혼합되면 오류가 발생하기 때문이다. 숫자만 있는 경우에는 필요없다.

5. 범주형 데이터 탐색

  - 칼럼 내의 고유한 값의 개수를 찾을 떄는 df.nunique()를 활용
  - 그 고유한 값이 무엇인지 알기 위해서는 df["칼럼명"].unique()를 활용
  - 항목별로 개수를 찾는 value_counts() 도 있다.

6. 기술통계량

  - int/float: .describe()를 통해 확인 가능
  - object: include파라미타를 추가적으로 적용
  > describe(include="O") # 대문자 O

# Section 8 인덱싱/슬라이싱(iloc)

iloc == integer location 정수 기반이다.(:로 슬라이싱하면 끝을 포함)

df.iloc[행번호, 열번호]

> 인덱스는 둘다 0부터시작

loc과 다른점은 숫자로 지정하는것만 차이가 있으며, :슬라이싱의 차이점 정도. 그리고 조건 추출이 안된다는 점 이 다르다.

# Section 11 필터링

필터링: 데이터프레임에서 조건을 적용해 원하는 데이터만 추출

> Boolean 값을 이용해서 조건을 설정한다.

조건문을 변수에 담아서 하는게 편하다.

> &: 교집합(and), |: 합집합(or)

> Not연산자: ~ / ex. ~cond

**.isin()**을 사용한 필터링: 괄호안의 **목록(list)**이 있는지 확인해 Boolean값 출력
> ex. df['메뉴'].isin(['녹차'])


```python
#dataframe 지정
df = pd.read_csv('cafe2.csv')
df
```

```python
#조건 1개
cond = df['칼로리']<50 #조건 지정 --> Boolean
df[cond] #조건에 부합하는 행출력
```

```python
#복수조건 필터링: 가격이 5000원 이상 & 칼로리가 100보다 큰 데이터
cond1 = df['가격']>=5000
cond2 = df['칼로리']> 100
df[cond1 & cond2]
```

```python
df[cond1 | cond2] #or 조건문
```

```python
# 원산지가 과테말라이면서 가격이 5000원 이하
cond1 = df['원산지'] == '과테말라'
cond2 = df['가격'] <=5000

df[cond1 & cond2]
```

```python
#isin()을 사용한 필터링
cond = df['메뉴'].isin(['녹차']) #boolean값 반환
df[cond]
```

```python
box = ['녹차','카푸치노','카페라떼']
cond = df['메뉴'].isin(box)
df[cond]
```

# Section 18 시계열 데이터(Timedelta)

datetime: 특정 시점의 날짜와 시간
Timedelta: 두 시점 사이의 차이 계산
- 날짜와 시간을 더하거나 빼서 새로운 시점을 계산할 수 있다.

- datetime자료형을 빼거다 더하면 timedelta 자료형이 된다
- total_seconds() 로 전체 초 구할 수 있음
- dt.days 혹은 dt.seconds로 시간이나 초만 볼 수 있음(합계X) 일 제외 시간을 초로 혹은 시간 제외 날짜 합계
- round(data)로 시간도 반올림 가능

```python
df = pd.read_csv('date_data.csv', usecols=['DateTime4'], parse_dates=['DateTime4'])
df
```

```python
# 특정 시간과의 차이
day = pd.Timedelta(days=99) #99일을 의미하게됨
df['100day'] = df['DateTime4'] + day
df
```

```python
# 시간
hour = pd.Timedelta(hours=100)
df['100hour'] = df['DateTime4'] + hour
df
```

```python
difference = pd.Timedelta(weeks=7, days=7, hours=7, minutes=7, seconds=7)
df['difference'] = df['DateTime4'] - difference
df
```

```python
# 두시간 사이의 차이
diff = df['100hour']- df['difference']
diff
```

```python
#timedelta 특정 값을 초 분 시간 일로 다음과 같이 나타낼 수 있음
#dt.total_seconds()

print(diff.dt.total_seconds()) #초
print(diff.dt.total_seconds()/60) #분
print(diff.dt.total_seconds()/60/60) #시간
print(diff.dt.total_seconds()/60/60/24) #일
```
